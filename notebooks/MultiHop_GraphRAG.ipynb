```python
import json
from pathlib import Path

# ------------ paths ---------------------------------------------------
json_path = Path("corpus.json")          # <-- your local file
txt_path  = Path("MultiHop/input/multihop_corpus.txt")

# ------------ load ----------------------------------------------------
with json_path.open(encoding="utf-8") as f:
    corpus = json.load(f)                # list[dict]

# ------------ write ---------------------------------------------------
with txt_path.open("w", encoding="utf-8") as out:
    for row in corpus:
        out.write(row["title"].strip() + "\n")
        meta = f'[{row["source"]} — {row["author"]} — {row["published_at"]}]'
        out.write(meta + "\n")
        out.write(row["body"].strip() + "\n\n")   # blank line between articles

print(f"✓ {len(corpus)} documents written to {txt_path.resolve()}")
```

```python
ROOT_DIR  = "/content/MultiHop"              # folder holding settings.yaml
# QUERY_JSON= "MultiHopRAG.json"          # your saved subset of the dataset
OUT_LOCAL = "ms_local.json"
OUT_GLOBAL= "ms_global.json"

GRAPH_CMD = ["python3", "-m", "graphrag.query", "--root", ROOT_DIR]

# optional: silence pandas warnings inside GraphRAG
import os, re, json, subprocess, tqdm, itertools
os.environ["PYTHONWARNINGS"] = "ignore::UserWarning"

# ─────────────────────────────────────────────────────────────────────────────
# 1.  helpers
# ─────────────────────────────────────────────────────────────────────────────
SUCCESS_RE = re.compile(r"SUCCESS: .*?Response:\s*(.*)", re.S)

def run_graphrag(method:str, question:str, timeout:int=1800) -> str:
    """Call GraphRAG CLI and return ONLY the answer string."""
    cmd = GRAPH_CMD + ["--method", method, question]
    proc = subprocess.run(
        cmd,
        stdout=subprocess.PIPE,
        stderr=subprocess.PIPE,
        text=True,
        timeout=timeout
    )
    # merge both streams – GraphRAG sometimes prints to stderr
    merged = proc.stdout + "\n" + proc.stderr
    m = SUCCESS_RE.search(merged)
    if not m:
        raise RuntimeError(f"Failed to parse answer for {method}: {question}\n"
                           f"CLI output was:\n{merged[:500]}")
    return m.group(1).strip()
```

```python
import json, random, math
from collections import defaultdict

# ------------------------------------------------------------------
# CONFIG -----------------------------------------------------------
QUERY_JSON        = "MultiHopRAG.json"            # full input file
OUTPUT_QUERY_JSON = "sampled_queries.json"    # 125-item subset
OUTPUT_INDEX_JSON = "sampled_indices.json"    # original indices
TARGET_TOTAL      = 125                       # desired sample size
RANDOM_SEED       = 42                        # reproducible seed
# ------------------------------------------------------------------

random.seed(RANDOM_SEED)

# 1. Load queries --------------------------------------------------
with open(QUERY_JSON, encoding="utf-8") as f:
    queries = json.load(f)

total_available = len(queries)
if TARGET_TOTAL > total_available:
    raise ValueError(f"Need {TARGET_TOTAL} questions but only {total_available} found.")

# 2. Bucket indices by question_type ------------------------------
type_to_indices = defaultdict(list)
for idx, q in enumerate(queries):
    type_to_indices[q["question_type"]].append(idx)

# 3. Compute how many to take from each type ----------------------
raw_target = {
    t: len(idxs) * TARGET_TOTAL / total_available
    for t, idxs in type_to_indices.items()
}

# floor counts and distribute remainder via “largest remainder”
sample_counts = {t: math.floor(v) for t, v in raw_target.items()}
leftover = TARGET_TOTAL - sum(sample_counts.values())
if leftover:
    remainders = {t: raw_target[t] - sample_counts[t] for t in raw_target}
    for t in sorted(remainders, key=remainders.get, reverse=True):
        if leftover == 0:
            break
        if sample_counts[t] < len(type_to_indices[t]):
            sample_counts[t] += 1
            leftover -= 1

# 4. Randomly pick the indices (seeded) ---------------------------
selected_indices = []
for t, n in sample_counts.items():
    selected_indices.extend(random.sample(type_to_indices[t], n))

# 5. Sort indices so output list follows original order -----------
selected_indices.sort()

# 6. Slice out the chosen query objects ---------------------------
sampled_queries = [queries[i] for i in selected_indices]

# 7. Dump results --------------------------------------------------
with open(OUTPUT_QUERY_JSON, "w", encoding="utf-8") as f:
    json.dump(sampled_queries, f, ensure_ascii=False, indent=2)

with open(OUTPUT_INDEX_JSON, "w", encoding="utf-8") as f:
    json.dump(selected_indices, f, indent=2)

print(f"Saved {len(sampled_queries)} queries → {OUTPUT_QUERY_JSON}")
print(f"Saved indices → {OUTPUT_INDEX_JSON}")
print("Counts per type:", {t: sample_counts[t] for t in sorted(sample_counts)})
```

```python
with open(OUTPUT_QUERY_JSON, encoding="utf-8") as f:
    queries = json.load(f)
```

```python
local_out, global_out = [], []
for row in tqdm.tqdm(queries, desc="GraphRAG"):
    q = row["query"]
    # ans_local  = run_graphrag("local",  q)
    try:
        ans_local  = run_graphrag("local",  q)
        ans_global = run_graphrag("global", q)
    except Exception as e:
        ans_local  = ans_global = f"[ERROR] {e}"

    # same schema as your earlier pipelines
    record = {
        "query"        : q,
        "model_answer" : None,          # filled below
        "gold_answer"  : row["answer"],
        "question_type": row["question_type"],
        "retrieval_list": []            # not evaluated – leave empty
    }
    lrec = record | {"model_answer": ans_local}
    grec = record | {"model_answer": ans_global}
    local_out.append(lrec)
    global_out.append(grec)
```

```python
# json.dump(local_out,  open(OUT_LOCAL,  "w"), ensure_ascii=False, indent=2)
json.dump(global_out, open(OUT_GLOBAL, "w"), ensure_ascii=False, indent=2)
print(f"✓ wrote {OUT_LOCAL} & {OUT_GLOBAL}")
```
